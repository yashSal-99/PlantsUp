```python
import os

def explore_directory(path, depth=0):
    # Print the current directory and its contents
    indent = "  " * depth  # Indentation for better visibility
    print(f"{indent}{os.path.basename(path)} ->")
    for item in os.listdir(path):
        item_path = os.path.join(path, item)
        if os.path.isdir(item_path):
            # If it's a folder, explore it recursively
            explore_directory(item_path, depth + 1)
        else:
            # If it's a file, print the file name
            print(f"{indent}  {item}")

# Set the root path
dataset_path = "/root/.cache/kagglehub/datasets/aryashah2k/indian-medicinal-leaves-dataset/versions/1"
explore_directory(dataset_path)
```

```python
import os
import shutil

# Define the paths for the two main datasets
dataset_1_path = "/root/.cache/kagglehub/datasets/aryashah2k/indian-medicinal-leaves-dataset/versions/1/Indian Medicinal Leaves Image Datasets/Medicinal plant dataset"
dataset_2_path = "/root/.cache/kagglehub/datasets/aryashah2k/indian-medicinal-leaves-dataset/versions/1/Indian Medicinal Leaves Image Datasets/Medicinal Leaf dataset"
output_path = "/content/Merged_Dataset"

# Create the output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)

# Function to merge photos
def merge_photos(folder_name):
    folder_1_path = os.path.join(dataset_1_path, folder_name)
    folder_2_path = os.path.join(dataset_2_path, folder_name)
    merged_folder_path = os.path.join(output_path, folder_name)

    # Create the merged folder
    os.makedirs(merged_folder_path, exist_ok=True)

    # Copy photos from both datasets
    if os.path.exists(folder_1_path):
        for file in os.listdir(folder_1_path):
            shutil.copy(os.path.join(folder_1_path, file), merged_folder_path)
    if os.path.exists(folder_2_path):
        for file in os.listdir(folder_2_path):
            shutil.copy(os.path.join(folder_2_path, file), merged_folder_path)

# Get unique folder names from both datasets
folders_1 = set(os.listdir(dataset_1_path))
folders_2 = set(os.listdir(dataset_2_path))
all_folders = folders_1.union(folders_2)

# Merge photos for each folder
for folder in all_folders:
    merge_photos(folder)

print(f"Photos merged successfully into: {output_path}")
```

```python
import os
import shutil
from google.colab import drive

# Mount Google Drive
# drive.mount('/content/drive')

# Path to the dataset in Google Drive
merged_dataset_path = "/content/drive/MyDrive/PlantsUp/Merged_Dataset"

# List of folder pairs to merge
folders_to_merge = [
    ("Tulasi", "Tulsi"),
    ("Nagadali", "Common rue(naagdalli)"),
    ("Ashoka", "ashoka"),  # If duplicates exist
    ("Brahmi", "Bhrami"),
    ("Curry", "Curry_Leaf"),
    ("Pomegranate", "Pomoegranate"),
    ("Guava", "Gauva"),
    ("Amruta_Balli", "Amruthaballi"),
    ("Lemon_grass", "Lemongrass"),
    ("Doddapatre", "Doddpathre"),
    ("Papaya", "Pappaya"),
    ("Palak(Spinach)", "Spinach1")
]

# Function to merge folders with conflict handling
def merge_folders(base_path, folder1, folder2):
    folder1_path = os.path.join(base_path, folder1)
    folder2_path = os.path.join(base_path, folder2)

    if os.path.exists(folder1_path) and os.path.exists(folder2_path):
        # Move contents of folder2 into folder1
        for item in os.listdir(folder2_path):
            source_path = os.path.join(folder2_path, item)
            destination_path = os.path.join(folder1_path, item)

            # Handle file name conflicts
            if os.path.exists(destination_path):
                base_name, ext = os.path.splitext(item)
                counter = 1
                new_name = f"{base_name}_{counter}{ext}"
                while os.path.exists(os.path.join(folder1_path, new_name)):
                    counter += 1
                    new_name = f"{base_name}_{counter}{ext}"
                destination_path = os.path.join(folder1_path, new_name)

            # Move the file
            shutil.move(source_path, destination_path)
        # Remove the now-empty folder2
        os.rmdir(folder2_path)
        print(f"Merged '{folder2}' into '{folder1}'")
    else:
        print(f"One or both folders do not exist: {folder1}, {folder2}")

# Perform merging directly in Drive
for folder1, folder2 in folders_to_merge:
    merge_folders(merged_dataset_path, folder1, folder2)

print("Folder merging complete!")
```
